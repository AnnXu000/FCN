```python
import os
import numpy as np
import matplotlib.pyplot as plt

from src.dataloader.dataloader import load_image
from src.dataloader.dataloader import VOCDataGenerator

poc_classes = ['bg', 'aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus', 'car', 'cat', 'chair', 'cow',\
                'diningtable', 'dog', 'horse', 'motorbike', 'person', 'pottedplant', 'sheep', 'sofa', 'train', 'tvmonitor']
print("Number of classes in VOC dataset: {}".format(len(poc_classes)))

def read_lines(filepath):
    with open(filepath, 'r') as file:
        return [line.strip() for line in file.readlines()]

def display_example(image, label):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title("Image")
    plt.imshow(image)
    plt.subplot(1, 2, 2)
    plt.title('Label')
    plt.imshow(label)
    plt.show()
```

```python
IMG_DIR = '../data/VOC2012/JPEGImages/'
LABEL_DIR = '../data/VOC2012/SegmentationClass/'
TRAIN_PATH = '../data/VOC2012/imagesets/Segmentation/train.txt'
VAL_PATH ='../data/VOC2012/ImageSets/Segmentation/val.txt' 
train_files = read_lines(TRAIN_PATH)
val_files = read_lines(VAL_PATH)
print("Number of training images: {}".format(len(train_files)))
print("Number of validation images: {}".format(len(val_files)))
```

```python
#list of path for training images and labels
train_img_files = [os.path.join(IMG_DIR, file+'.jpg') for file in train_files]
train_label_files = [os.path.join(LABEL_DIR, file+'.png') for file in train_files]

img, label = load_image(train_img_files[0]), load_image(train_label_files[0])
display_example(img, label)
```

```python
#create generator for training images and labels
train_generator = VOCDataGenerator(train_img_files, train_label_files, img_size=(320, 320))
```

```python
from time import time
start = time()
for i in range(len(train_generator)):
    imgs, labels = train_generator[i]
    print(imgs.shape, labels.shape)
stop = time()    
print("Time taken: {:.4f}".format(stop-start))    
```
